```python
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
import geopandas as gpd
import seaborn as sbn
```

```python
# Definimos los nombres de los archivos y levantamos los datasets
path = ''

argenprop = 'data_argenprop.csv'
meli = 'data_meli.csv'
zonaprop = 'zonaprop_procesado.csv'
properati = 'properati.csv'
barrios = 'barrios'

d_a = pd.read_csv(path + argenprop).dropna(subset = ['precio'])
d_z = pd.read_csv(path + zonaprop).dropna(subset = ['precio'])
d_p = pd.read_csv(path + properati).dropna(subset = ['precio'])
d_m = pd.read_csv(path + meli).dropna(subset = ['precio'])

def fuente(d, fuente_d):
    if 'fuente' not in d.columns:
        d['fuente'] = fuente_d
    try:
        return d.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis = 1)
    except:
        return d
d_a = fuente(d_a, 'argenprop')
d_z = fuente(d_z, 'zonaprop')
d_p = fuente(d_p, 'properati')
d_m = fuente(d_m, 'meli')
```

```python
d_a = pd.read_csv(path + argenprop).dropna(subset = ['precio'])
d_z = pd.read_csv(path + zonaprop).dropna(subset = ['precio'])
d_p = pd.read_csv(path + properati).dropna(subset = ['precio'])

d_z['fecha'] = pd.to_datetime(d_z.fecha)
d_p['fecha'] = pd.to_datetime(d_p.fecha)


# Filtremos por fecha (zonaprop y properati)

fecha_desde = pd.to_datetime('2021-01-01')
d_z = d_z[d_z.fecha > fecha_desde]
d_p = d_p[d_p.fecha > fecha_desde]

# Filtramos propiedades en dolares

d_p = d_p[d_p.precio_moneda.str.contains('ars', case = False)]
d_a = d_a[(d_a.moneda == '$')]
d_z = d_z[d_z.precio_moneda == '$']

# Filtramos propiedades por precio 

d_p = d_p[d_p.precio < d_p.precio.quantile(0.9)]
d_a = d_a[d_a.precio < d_a.precio.quantile(0.9)]
d_z = d_z[d_z.precio < d_z.precio.quantile(0.9)]
# Creemos una columna sup_mt2

d_a['sup_mt2'] = d_a.precio / d_a.sup_total
d_p['sup_mt2'] = d_p.precio / d_p.sup_total
d_z['sup_mt2'] = d_z.precio / d_z.iconArea

d_barrios = gpd.read_file(path + barrios)
d_barrios.to_crs("EPSG:4326", inplace = True)
```

```python
d_z.sup_mt2.dropna()
```

```python
fig, ax = plt.subplots(1, 1, figsize = (10,24))
ax.set_title('Argenprop')
d_barrios.plot(ax = ax)
sbn.scatterplot(data = d_a[d_a.precio < 1000000],
                x = 'longitud',
         y = 'latitud',
         ax = ax,
         hue = 'sup_mt2',
         cmap = 'Reds',)
ax.set_xlim(-58.55,-58.3)
ax.set_ylim(-34.75,-34.5)
plt.show()

fig, ax = plt.subplots(1, 1, figsize = (10,24))
ax.set_title('Properati')
d_barrios.plot(ax = ax)
sbn.scatterplot(data = d_p,
         x = 'lon',
         y = 'lat',
         ax = ax,
         hue = d_p['sup_mt2'],
         cmap = 'Reds',
         )
ax.set_xlim(-58.55,-58.3)
ax.set_ylim(-34.75,-34.5)
plt.show()

fig, ax = plt.subplots(1, 1, figsize = (10,24))
ax.set_title('Zonaprop')
d_barrios.plot(ax = ax)
sbn.scatterplot(data = d_z,
         x = 'lon',
         y = 'lat',
         ax = ax,
         hue = 'sup_mt2',
         cmap = 'Reds',)
ax.set_xlim(-58.55,-58.3)
ax.set_ylim(-34.75,-34.5)

plt.show()
```
