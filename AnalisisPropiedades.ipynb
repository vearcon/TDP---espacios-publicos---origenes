```python
import pandas as pd
import matplotlib.pylab as plt
import numpy as np
import geopandas as gpd
import seaborn as sbn
from scipy import stats
```

```python
# Definimos los nombres de los archivos y levantamos los datasets
path = ''

argenprop = 'data_argenprop.csv'
meli = 'data_meli.csv'
zonaprop = 'zonaprop_procesado.csv'
properati = 'properati.csv'
barrios = 'barrios'

d_a = pd.read_csv(path + argenprop).dropna(subset = ['precio'])
d_z = pd.read_csv(path + zonaprop).dropna(subset = ['precio'])
d_p = pd.read_csv(path + properati).dropna(subset = ['precio'])
d_m = pd.read_csv(path + meli).dropna(subset = ['precio'])

def fuente(d, fuente_d):
    if 'fuente' not in d.columns:
        d['fuente'] = fuente_d
    try:
        return d.drop(['Unnamed: 0', 'Unnamed: 0.1'], axis = 1)
    except:
        return d
def fecha(d, f_min = '2021-01-01'):
    d['fecha'] = pd.to_datetime(d['fecha'], utc = 'UTC')
    return d[d.fecha > pd.to_datetime(f_min, utc = 'UTC')]

d_a = fuente(d_a, 'argenprop')
d_z = fuente(d_z, 'zonaprop')
d_p = fuente(d_p, 'properati')
d_m = fuente(d_m, 'meli')

#d_a = fecha(d_a)
d_z = fecha(d_z)
d_p = fecha(d_p)
d_m = fecha(d_m)

# Columnas a Trabajar

columnas = ['lat',
            'lon',
            'precio',
            'sup_total',
            'expensas',
            'ambientes',
            'fuente',
            'precio_moneda'
            ]
d_a = d_a[columnas]
d_z = d_z[columnas]
d_p = d_p[columnas]
d_m = d_m[columnas]

d_a = d_a[(d_a.precio_moneda == '$') | (d_a.precio_moneda == 'ars')]
d_z = d_z[(d_z.precio_moneda == '$') | (d_z.precio_moneda == 'ars')]
d_p = d_p[(d_p.precio_moneda == '$') | (d_p.precio_moneda == 'ars')]
d_m = d_m[(d_m.precio_moneda == '$') | (d_m.precio_moneda == 'ars')]

d_a = d_a.dropna(subset = set(columnas) - {'expensas'}).fillna(0)
d_z = d_z.dropna(subset = set(columnas) - {'expensas'}).fillna(0)
d_p = d_p.dropna(subset = set(columnas) - {'expensas'}).fillna(0)
d_m = d_m.dropna(subset = set(columnas) - {'expensas'}).fillna(0)

# Fracciones de datos analizados
print(f'Argenprop : {round(len(d_a) / len(pd.read_csv(path + argenprop).dropna(subset = ["precio"])),4)}')
print(f'Zonaprop : {round(len(d_z) / len(pd.read_csv(path + zonaprop).dropna(subset = ["precio"])),4)}')
print(f'Properati : {round(len(d_p) / len(pd.read_csv(path + properati).dropna(subset = ["precio"])),4)}')
print(f'MercadoLibre : {round(len(d_m) / len(pd.read_csv(path + meli).dropna(subset = ["precio"])),4)}')
```

```python
# Hacemos una inspección de lo que tenemos

print(d_a.head())
print(d_z.head())
print(d_p.head())
print(d_m.head())

d = d_a.append(d_z).append(d_p).append(d_m)
```

```python
# El data set unificado
d = gpd.GeoDataFrame(d,
                     geometry = gpd.points_from_xy(y = d['lat'],
                                                   x = d['lon']
                                                  ),
                     crs = "EPSG:4326"
                    )
# Los polígonos de barrios
d_barrios = gpd.read_file(path + barrios).to_crs("EPSG:4326")

# Filtramos con los que están en CABA
d = gpd.sjoin(d, d_barrios[['BARRIO', 'geometry']])
```

```python
sbn.boxplot(data = d, y = 'precio', x = 'fuente')
# Se ve que hay outliers en el precio que no deben corresponder a dpts reales
```

```python
# Filtro en precio, sacando outliers
d = d[(d.precio < d.precio.quantile(0.99)) & (d.precio > d.precio.quantile(0.01))]
d = d[(d.sup_total < d.sup_total.quantile(0.99)) & (d.sup_total > d.sup_total.quantile(0.01))]
```

```python
sbn.boxplot(data = d, y = d['precio'] / d['sup_total'], x = 'fuente')
```

```python
fig, ax = plt.subplots(figsize = (15,15))

d_barrios.plot(ax = ax, color = 'grey', alpha = 0.5)

sbn.scatterplot(data = d,
                y ='lat',
                x = 'lon', 
                style = 'ambientes',
                hue = np.log(d['precio'] / d['sup_total']),
                size = d['sup_total'],
                ax = ax,
                palette = 'magma'
               )

ax.axis('equal')

plt.show()
```

```python
# Análisis agregado
# - distribución ambientes
# - metro cuadrado
# - barrios
```

```python
fig, ax = plt.subplots(figsize = (10,8))
ax.set_title('Oferta de alquileres según N° de ambientes',
            fontsize = 15)
d.ambientes.value_counts().sort_index().plot(kind = 'bar', ax = ax)
ax.grid('on', linestyle = 'dashed')
ax.set_ylabel('Cantidad de Propiedades')
ax.set_xlabel('Cantidad de Ambientes')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10,8))
ax.set_title('Distribución de la Superficie Total',
            fontsize = 15)
d.sup_total.hist(ax = ax,
                 bins = 25,
                 label = f'La media es: {round(d.sup_total.mean(),3)} $m^2$')
ax.grid('on', linestyle = 'dashed')
ax.set_ylabel('Cantidad de Propiedades')
ax.set_xlabel('Cantidad de $m^2$')
ax.legend()
plt.show()
```

```python
# Análisis por barrio
d['precio_superficie'] = d['precio'] / d['sup_total']
```

```python
fig, ax = plt.subplots(figsize = (10,8))
ax.set_title('Distribución del Precio por $m²2',
            fontsize = 15)
d.precio_superficie.hist(ax = ax, bins = 25, label = f'La media es: {round(d.precio_superficie.mean(),3)} \$')
ax.grid('on', linestyle = 'dashed')
ax.set_ylabel('Cantidad de Propiedades')
ax.set_xlabel('Precio del $m^2$ [\$]')
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10,10))
ax.set_title('Cantidad de Propiedades por Barrio', fontsize = 15)
(d.groupby('BARRIO').precio.count().sort_values(ascending = True) / d.groupby('BARRIO').precio.count().sort_values(ascending = True).sum()).plot(kind = 'barh',
                                                                      ax = ax)
ax.grid('on',
        linestyle = 'dashed')
plt.show()
```

```python
d_barrios = d_barrios.merge(d.groupby('BARRIO').precio_superficie.mean(), on = 'BARRIO')
```

```python
fig, ax = plt.subplots(figsize = (15,15))

im = d_barrios.plot(ax = ax,
                    column = 'precio_superficie',
                    vmax = d_barrios.precio_superficie.quantile(0.95),
                    vmin = d_barrios.precio_superficie.quantile(0.05),
                    legend = True)


ax.axis('equal')
ax.axis('off')
ax.set_title('Precio por $m^2$ Medio por Barrio',
            fontsize = 15)
plt.show()
```
